A Common Table Expression (CTE) is a temporary named result set that you can reference within a single SELECT, INSERT, UPDATE, or DELETE statement. Defined using the WITH keyword, a CTE acts like a virtual table that exists only for the duration of that specific query execution.

CTEs dramatically enhance code readability by replacing complex, deeply nested subqueries with a clean, top-to-bottom logical structure

Views in SQL are considered updatable when the database can unambiguously trace every row and column in the view back to its exact source row in the underlying base table. If the database cannot uniquely map the changes (e.g., inserts, updates, deletes) back to a single source table, the view becomes read-only.

ANS Q3--

Using stored procedures instead of repeating raw SQL queries offers significant advantages across performance, security, and code maintenanc.

**Enhanced Performance & Efficiency**
Execution Plan Caching: The database compiles and optimizes a stored procedure once upon its initial run. Subsequent calls reuse this cached execution plan, skipping the parsing and optimization phases required by raw queries.

**Drastically Reduced Network Traffic**: Instead of passing hundreds of lines of raw SQL text over the network repeatedly, your application only sends the short procedure name and its parameter values.

**Server-Side Processing: **Complex, multi-step operations are executed entirely within the database engine. This prevents large intermediate datasets from being transferred back and forth to your application.

**Robust SecuritySQL Injection Mitigation:** Stored procedures inherently separate the query logic from the data by using strictly typed input parameters. This makes it virtually impossible for malicious user input to alter the executable command structure.

Triggers are automated database programs that execute ("fire") in response to specific events, such as inserting, updating, or deleting data. Their primary purpose is to enforce complex business rules, ensure data integrity, and automate routine tasks without requiring intervention from the application or use.

Triggers are essential for audit logging in applications where tracking historical changes is required for security or regulatory compliance (e.g., financial or healthcare databases.
